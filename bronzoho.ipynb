```python
import requests

CLIENT_ID     = "1000.36KIEIGAC79U7PB0Y5IA74IDRFVT0M"
CLIENT_SECRET = "ecd0a7ff65710a51a63306894821b405a9bb66c489"
REFRESH_TOKEN = "1000.82d7700d7767920dff17a5720274e143.f8e0a8e33fb818a16bcd7b70e1117892"

# shape checks (safe to share output)
print("client_id len:", len(CLIENT_ID), "| starts 1000.:", CLIENT_ID.startswith("1000."))
print("secret len:", len(CLIENT_SECRET))
print("refresh dots:", REFRESH_TOKEN.count("."), "| len:", len(REFRESH_TOKEN), "| has ellipsis:", "…" in REFRESH_TOKEN)
print("-" * 40)

r = requests.post(
    "https://accounts.zoho.in/oauth/v2/token",
    data={
        "grant_type": "refresh_token",
        "client_id": CLIENT_ID,
        "client_secret": CLIENT_SECRET,
        "refresh_token": REFRESH_TOKEN,
    },
    timeout=30,
)
print("status:", r.status_code)
print("body:", r.text[:300])
```

```python
tok = r.json()
access_token = tok["access_token"]
api_domain   = tok["api_domain"]

leads = requests.get(
    f"{api_domain}/crm/v8/Leads",
    headers={"Authorization": f"Zoho-oauthtoken {access_token}"},
    params={
        "fields": "Last_Name,First_Name,Email,Company,Lead_Status,Modified_Time",
        "per_page": 200,
    },
    timeout=60,
)
print("status:", leads.status_code)

if leads.status_code == 200:
    data = leads.json().get("data", [])
    print("lead count:", len(data))
    for ld in data[:3]:
        print(ld.get("id"), "|", ld.get("Last_Name"), "|", ld.get("Email"))
else:
    print("body:", leads.text[:300])
```
