```python
import os, shutil
from datetime import datetime
L, P = "/lakehouse/default/Files/bronze/landing", "/lakehouse/default/Files/bronze/processed"
for folder in ["agent","builder","customer","expense","leads","lease","property","sales"]:
    src = f"{L}/{folder}"
    if not os.path.isdir(src): continue
    os.makedirs(f"{P}/{folder}", exist_ok=True)
    for fn in os.listdir(src):
        stem, ext = os.path.splitext(fn)
        shutil.move(f"{src}/{fn}", f"{P}/{folder}/{stem}__{datetime.now():%Y%m%d_%H%M%S}{ext}")
print("archived")
```
